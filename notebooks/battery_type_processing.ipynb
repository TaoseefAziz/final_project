```python
import pandas as pd
from utils import regex

SPECIFICATIONS_FILE = "../specifications.csv"
```

# loading the file

```python
raw_specs = pd.read_csv(SPECIFICATIONS_FILE)

raw_specs
```

```python
battery_types = raw_specs["Battery_Type"]

battery_types
```

```python
total_energy_regex = r"(\d+) mAh"

battery_life = battery_types.map(regex.extractor_generator(total_energy_regex))

converted = raw_specs.assign(total_battery=battery_life).drop(columns=["Battery_Type"]).dropna(subset=["total_battery"])

converted
```
